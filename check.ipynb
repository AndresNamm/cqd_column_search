```python
import pandas as pd

schema= pd.read_csv('schema.csv')
docs = pd.read_csv('docs_schema.csv')
measures = pd.read_csv('measures_schema.csv')

measures

```

```python
res = schema.merge(docs, on='Name', how='left').merge(measures, on='Name', how='left')
res['Description']= res['Description_x'].fillna(res['Description_y'])
```

```python
res[['Type', 'Category', 'Name', 'Description', 'Units','Possible Reasons for blank values']].to_csv('updated_schema.csv', index=False)
```
